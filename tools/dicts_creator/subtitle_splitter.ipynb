```python
all_words_filename = "all_words.txt"
with open(all_words_filename, "r", encoding="utf-8") as f:
    all_words = sorted(set([w.lower() for w in f.read().split() if w]))
    
all_words
```

```python
sub_filename = "Puss-in-Boots -The-Last-Wish_2022_Dutch.srt"

with open(sub_filename, "r", encoding="utf-8") as f:
    sub_raw = f.read()
    
sub_words = sorted(set([word.lower() for word in sub_raw.split() if word.isalpha()]))
len(sub_words)
```

```python
sub_words_filtered = [w for w in sub_words if not w in all_words]
len(sub_words_filtered)
```

```python
out_words_filename = "puss-in-boots-2-Dutch-sub-words.txt"
with open(out_words_filename, "w") as f:
    f.write("\n".join(sub_words_filtered))
    f.write("\n")
```
