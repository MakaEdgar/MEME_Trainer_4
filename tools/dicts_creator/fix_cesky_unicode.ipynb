```python
new_words = []
for w in words:
    w_new = w.encode()
    w_new = w_new.replace(b'e\xcc\x8c', b'\xc4\x9b')     # wrong ě
    w_new = w_new.replace(b's\xcc\x8c', b'\xc5\xa1')     # wrong š
    w_new = w_new.replace(b'c\xcc\x8c', b'\xc4\x8d')     # wrong č
    w_new = w_new.replace(b'r\xcc\x8c', b'\xc5\x99')     # wrong ř
    w_new = w_new.replace(b'z\xcc\x8c', b'\xc5\xbe')     # wrong ž
    w_new = w_new.replace(b'y\xcc\x81', b'\xc3\xbd')     # wrong ý
    w_new = w_new.replace(b'a\xcc\x81', b'\xc3\xa1')     # wrong á
    w_new = w_new.replace(b'i\xcc\x81', b'\xc3\xad')     # wrong í
    w_new = w_new.replace(b'u\xcc\x81', b'\xc3\xba')     # wrong ú
    w_new = w_new.replace(b'u\xcc\x8a', b'\xc5\xaf')     # wrong ů
    w_new = w_new.decode()
    new_words.append(w_new)
    print(w_new)
```
